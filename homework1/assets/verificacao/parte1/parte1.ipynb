# Exercise 1

Compute the correlation (Pearson correlation) and Spearman's rank for two variables x1 and x2=f(x1). Are the values the same or different, please justify. (Indicate all computational steps!) The functions correspond to activation functions, that will be later introduced when dealing with artificial neural networks.

```python
import pearson as p
import spearman as s
import math
```

## a) x1 = (−3,−2,−1,0,1,2), x2 = 0.5*x1

Clica [aqui](../../resolucao/parte1/a\).png) para ver a resolução

```python
n = 6
x1 = [-3,-2,-1,0,1,2]
x2 = []

for x in x1:
    x2.append(x*0.5)

pearsonValue = p.calculate_pearson(n,x1,x2)
spearmanValue = s.calculate_spearman(n,x1,x2)

print("pearson",pearsonValue)
print("spearman",spearmanValue)
```

## b) x1 = (−3,−2,−1,0,1,2), x2 = max(0,x1)
Clica [aqui](../../resolucao/parte1/b\_parte1.png) para ver a 1ª parte da resolução;
Clica [aqui](../../resolucao/parte1/b\_parte2.png) para ver a 2ª parte da resolução

```python
n = 6
x1 = [-3,-2,-1,0,1,2]
x2 = []

for x in x1:
    x2.append(max(0,x))

pearsonValue = p.calculate_pearson(n,x1,x2)
spearmanValue = s.calculate_spearman(n,x1,x2)

print("pearson",pearsonValue)
print("spearman",spearmanValue)
```

## c) x1 = (−3,−2,−1,0,1,2), x2 = 1/ (1+ e^(-x))
Clica [aqui](../../resolucao/parte1/c\_parte1.png) para ver a 1ª parte da resolução;
Clica [aqui](../../resolucao/parte1/c\_parte2.png) para ver a 2ª parte da resolução

```python
n = 6
x1 = [-3,-2,-1,0,1,2]
x2 = []

for x in x1:
    numerator = 1
    denominator = 1 + math.exp(-x)
    x2.append(numerator/denominator)

pearsonValue = p.calculate_pearson(n,x1,x2)
spearmanValue = s.calculate_spearman(n,x1,x2)

print("pearson",pearsonValue)
print("spearman",spearmanValue)
```
